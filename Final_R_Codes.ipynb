Read Data

```r
# Read the CSV file
mean <- read.csv("Final_mean.csv")
sd <- read.csv("Final_sd.csv")
skewness <- read.csv("Final_skewness.csv")
kurtosis <- read.csv("Final_kurtosis.csv")
mean$date <- as.Date(mean$Date, format="%Y/%m/%d")
sd$date <- as.Date(sd$Date, format="%Y/%m/%d")
skewness$date <- as.Date(skewness$Date, format="%Y/%m/%d")
kurtosis$date <- as.Date(kurtosis$Date, format="%Y/%m/%d")
head(sd)
```

Function to calculate and plot wavelet coherence

```r
wavelet_coherence_plot <- function(s1, data, col1, col2, nrand = 100) {
  n <- nrow(data)
  t1 <- cbind(1:n, data[[col1]])
  t2 <- cbind(1:n, data[[col2]])
  dates <- mean$date
  # Compute wavelet coherence
  wtc.AB <- wtc(t1, t2, nrands = nrand)

  # Create a sequence of monthly dates
  years <- format(dates, "%Y")
  unique_years <- unique(years)
  at <- match(unique_years, years) - 1
  year_labels <- unique_years

  # Plot wavelet coherence with date labels on the x-axis
  par(oma = c(0, 0, 0, 1), mar = c(5, 4, 5, 5) + 0.1)
  plot(wtc.AB, plot.phase = TRUE, xaxt = 'n', lty.coi = 1, col.coi = "black",
       lwd.coi = 2, lwd.sig = 2, arrow.lwd = 0.05, arrow.len = 0.08,
       ylab = "Scale", xlab = "Year", plot.cb = TRUE, main = paste(s1, "movement wavelet coherence between", col1, "and", col2))

  axis(side = 1, at = at, labels = year_labels)
}

# Example usage:
# Assuming data1 is your dataframe and columns "BTC" and "Energy" exist
# wavelet_coherence_plot(data1, "BTC", "Energy")
```

Plot mean coherence heatmaps

```r
wavelet_coherence_plot("First", mean, "BTC", "Energy")
wavelet_coherence_plot("First", mean, "BTC", "GPR")
wavelet_coherence_plot("First", mean, "Energy", "GPR")
```

Plot standard deviation coherence heatmaps

```r
wavelet_coherence_plot("Second", sd, "BTC", "Energy")
wavelet_coherence_plot("Second", sd, "BTC", "GPR")
wavelet_coherence_plot("Second", sd, "Energy", "GPR")
```

Plot skewness coherence heatmaps

```r
wavelet_coherence_plot("Third", skewness, "BTC", "Energy")
wavelet_coherence_plot("Third", skewness, "BTC", "GPR")
wavelet_coherence_plot("Third", skewness, "Energy", "GPR")
```

Plot kurtosis coherence heatmaps

```r
wavelet_coherence_plot("Fourth", kurtosis, "BTC", "Energy")
wavelet_coherence_plot("Fourth", kurtosis, "BTC", "GPR")
wavelet_coherence_plot("Fourth", kurtosis, "Energy", "GPR")
```
